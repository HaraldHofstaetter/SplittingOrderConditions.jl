```julia
using SplittingOrderConditions
```

```julia
p,s = 6,10
@time begin
    eqs = generate_equations(1,s, symmetric=true, with_brackets=false)
    for q=2:p
        merge!(eqs,generate_equations(q,s, symmetric=true, with_brackets=false)) 
    end
end    
eqs = [string(eq,";") for (lw,eq) in eqs];
```

```julia
using PyCall
@pyimport phcpy
@pyimport phcpy.solver as solver
```

```julia
@time sols = solver.solve(eqs, tasks=4)
```

```julia
@pyimport phcpy.solutions as solutions
```

```julia
eqs_lem = generate_equations(p+1,s)
```

```julia
lem_eq = join([string("abs(",replace(replace(eq,"(", "["), ")", "]"),")^2") for eq in values(eqs_lem)], :+)
@time lem = eval(parse(string( "(a,b)->sqrt( begin ", lem_eq, " end )")))
```

```julia
vars = ["a1", "a2", "a3", "a4", "a5", "b1", "b2", "b3", "b4", "b5"]
lem_min_real = 1e10
lem_min_real_1 = 1e10
lem_min_pos_cmplx = 1e10
idx_min_real = -1
idx_min_real_1 = -1
idx_min_pos_cmplx = -1
idx = 0
a=[0]
b=[0]
for sol in sols
    soldic = solutions.strsol2dict(sol)
    coeffs = Complex128[soldic[var] for var in vars]
    idx += 1
    i = maximum(abs(imag(coeffs)))
    if i<1e-12
        coeffs = Float64[real(x) for x in coeffs]
    end    
    a = vcat(coeffs[1:5],reverse(coeffs[1:5]))
    b = vcat(coeffs[6:10],reverse(coeffs[6:9]),0.0)
    n = lem(a,b)
    print("i=", idx, " lem=", n)
    if i<1e-12 
        print(" R")
        if n< lem_min_real
            lem_min_real = n
            idx_min_real = idx
            print(" MIN")
        end
        if all(abs(a).<=ones(s)) && all(abs(b).<=ones(s)) && n< lem_min_real_1
            lem_min_real_1 = n
            idx_min_real_1 = idx
            print(" 1")
        end
    elseif  minimum(real(coeffs)) > 0 
        print(" P")
        if n< lem_min_pos_cmplx
            lem_min_pos_cmplx = n
            idx_min_pos_cmplx = idx
            print(" MIN")
        end
    end
    println()
end    
```

```julia
println("MIN_REAL (idx=", idx_min_real,", lem=", lem_min_real, "):")  
soldic = solutions.strsol2dict(sols[idx_min_real])
coeffs = Float64[real(soldic[var]) for var in vars]
a = vcat(coeffs[1:5],reverse(coeffs[1:5]))
b = vcat(coeffs[6:10],reverse(coeffs[6:9]),0.0)
println(a) 
println(b)
println()
println("MIN_REAL_1 (idx=", idx_min_real_1,", lem=", lem_min_real_1, "):")  
soldic = solutions.strsol2dict(sols[idx_min_real_1])
coeffs = Float64[real(soldic[var]) for var in vars]
a = vcat(coeffs[1:5],reverse(coeffs[1:5]))
b = vcat(coeffs[6:10],reverse(coeffs[6:9]),0.0)
println(a) 
println(b)
println()
println("MIN_CMPLX_POSITIVE_REALPART (idx=", idx_min_pos_cmplx,", lem=", lem_min_pos_cmplx ,"):")   
soldic = solutions.strsol2dict(sols[idx_min_pos_cmplx])
coeffs = Complex128[soldic[var] for var in vars]
a = vcat(coeffs[1:5],reverse(coeffs[1:5]))
b = vcat(coeffs[6:10],reverse(coeffs[6:9]),0.0)
println(a) 
println(b)
```
