```julia
using SplittingOrderConditions
```

```julia
p,s = 5,8
@time begin
    eqs = generate_equations(1,s, palindromic=true, with_brackets=false)
    for q=2:p
       merge!(eqs,generate_equations(q,s, palindromic=true, with_brackets=false)) 
    end
end    
eqs = [string(eq,";") for (lw,eq) in eqs]
```

```julia
using PyCall
```

```julia
@pyimport phcpy
```

```julia
@pyimport phcpy.solver as solver
```

```julia
@time sols = solver.solve(eqs, tasks=4)
```

```julia
@pyimport phcpy.solutions as solutions
```

```julia
eqs_lem = generate_equations(p+1,s)
```

```julia
lem_eq = join([string("abs(",replace(replace(eq,"(", "["), ")", "]"),")^2") for eq in values(eqs_lem)], :+)
@time lem = eval(parse(string( "(a,b)->sqrt( begin ", lem_eq, " end )")))
```

```julia
vars = ["a1", "a2", "a3", "a4", "b1", "b2", "b3", "b4"]
lem_min_real = 1e10
lem_min_real_1 = 1e10
lem_min_pos_cmplx = 1e10
idx_min_real = -1
idx_min_real_1 = -1
idx_min_pos_cmplx = -1
idx = 0
a=[0]
b=[0]
for sol in sols
    soldic = solutions.strsol2dict(sol)
    coeffs = Complex128[soldic[var] for var in vars]
    idx += 1
    i = maximum(abs(imag(coeffs)))
    if i<1e-12
        coeffs = Float64[real(x) for x in coeffs]
    end    
    a = vcat(coeffs[1:4],reverse(coeffs[5:8]))
    b = reverse(a) 
    n = lem(a,b)
    print("i=", idx, " lem=", n)
    if i<1e-12 
        print(" R")
        if n< lem_min_real
            lem_min_real = n
            idx_min_real = idx
            print(" MIN")
        end
        if all(abs(a).<=ones(8)) && all(abs(b).<=ones(8)) && n< lem_min_real_1
            lem_min_real_1 = n
            idx_min_real_1 = idx
            print(" 1")
        end
    elseif  minimum(real(coeffs)) > 0 
        print(" P")
        if n< lem_min_pos_cmplx
            lem_min_pos_cmplx = n
            idx_min_pos_cmplx = idx
            print(" MIN")
        end
    end
    println()
end    
```
